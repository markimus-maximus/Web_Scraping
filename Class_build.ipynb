```python
import requests

from bs4 import  BeautifulSoup
import pandas as pd
```

```python
#creating empty lists with which to store data later
crypto_name_list = []
crypto_market_cap_list = []
crypto_price_list = []
crypto_circulating_supply_list = []
crypto_ticker_list = []

```

```python
#creating a data frame for later
df = pd.DataFrame()
```

```python
#create function to scrape data
#example = https://coinmarketcap.com/historical/20220703/'
def crypto_scrape(date = '20220703/'):
    #define url and vary depending on the date
    url = 'https://coinmarketcap.com/historical/'+date
    #request webpage
    webpage = requests.get(url)
    #parse the html
    soup = BeautifulSoup(webpage.text, 'html.parser')
    #create a variable for the table rows (tr)
    tr = soup.find_all('tr', attrs={'class':'cmc-table-row'})
    #creating a counter to iterate through first 10 rows (top 10 cryptocurrencies)
    count = 0
    #iterate through the first 10 rows
    for row in tr:
        if count == 10:
            break
        count = count + 1

    #store the name of the coin as a variable by finding td element (column) 
        name_column = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__name'})
        #find the column within the row and strip the text
        crypto_name = name_column.find('a', attrs={'class': 'cmc-table__column-name--name cmc-link'}).text.strip()
        print(crypto_name)

        #find and createvvariable for market cap
        crypto_market_cap = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap'}).text.strip()
        print(crypto_market_cap)

        #Find and create variable for the price
        crypto_price = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price'}).text.strip()
        print(crypto_price)
        #Find and create variable for the circulating supply
        crypto_circulating_supply_and_ticker = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply'}).text.strip()
        print(crypto_circulating_supply_and_ticker)
        #Find and create variable for the circulating supply by splitting the supply+ticker variable
        crypto_ticker = crypto_circulating_supply_and_ticker.split(' ')[1]
        print(crypto_ticker)
        #Find and create variable for the coin ticker
        crypto_circulating_supply = crypto_circulating_supply_and_ticker.split(' ')[0]
        print(crypto_circulating_supply)
        

        #append the data to the associated lists
        crypto_name_list.append(crypto_name)
        crypto_market_cap_list.append(crypto_market_cap)
        crypto_price_list.append(crypto_price)
        crypto_circulating_supply_list.append(crypto_circulating_supply)
        crypto_ticker_list.append(crypto_ticker)


```

```python
#logic for determining date
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2013, 4, 28)
end_date = date(2022, 7, 3)
for single_date in daterange(start_date, end_date):

    print(single_date.strftime("%Y%m%d"))
```

```python
#run the function with a nominated date
crypto_scrape(date = '20220703/')
```

```python
#store the data in a dataframe to make it structured
df['Name'] = crypto_name_list
df['Market cap'] = crypto_market_cap_list
df['Price'] = crypto_price_list
df['Circulating supply'] = crypto_circulating_supply_list
df['Ticker'] = crypto_ticker_list

#show the data
df
```
