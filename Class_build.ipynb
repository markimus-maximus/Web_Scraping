```python
#create function to scrape data
#example = https://coinmarketcap.com/historical/20220703/'
def crypto_scrape():
    #creating empty lists with which to store data later
    crypto_name_list = []
    crypto_market_cap_list = []
    crypto_price_list = []
    crypto_circulating_supply_list = []
    crypto_ticker_list = []
    
    #creating a data frame for later
    df = pd.DataFrame()
    
    #define url and vary depending on the date
    url = 'https://coinmarketcap.com/historical/20220703'
    
    #request webpage
    webpage = requests.get(url)
    
    #parse the html
    soup = BeautifulSoup(webpage.text, 'html.parser')
   
    #create a variable for the table rows (tr)
    tr = soup.find_all('tr', attrs={'class':'cmc-table-row'})
    
    #creating a counter to iterate through first 10 rows (top 10 cryptocurrencies)
    count = 0
    
    #iterate through the first 10 rows
    for row in tr:
        if count == 10:
            break
        count = count + 1

    #store the name of the coin as a variable by finding td element (column) 
        name_column = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__name'})
        #find the column within the row and strip the text
        crypto_name = name_column.find('a', attrs={'class': 'cmc-table__column-name--name cmc-link'}).text.strip()
        print(crypto_name)

        #find and createvvariable for market cap
        crypto_market_cap = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap'}).text.strip()
        print(crypto_market_cap)

        #Find and create variable for the price
        crypto_price = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price'}).text.strip()
        print(crypto_price)
        #Find and create variable for the circulating supply
        crypto_circulating_supply_and_ticker = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply'}).text.strip()
        print(crypto_circulating_supply_and_ticker)
        #Find and create variable for the circulating supply by splitting the supply+ticker variable
        crypto_ticker = crypto_circulating_supply_and_ticker.split(' ')[1]
        print(crypto_ticker)
        #Find and create variable for the coin ticker
        crypto_circulating_supply = crypto_circulating_supply_and_ticker.split(' ')[0]
        print(crypto_circulating_supply)
        

        #append the data to the associated lists
        crypto_name_list.append(crypto_name)
        crypto_market_cap_list.append(crypto_market_cap)
        crypto_price_list.append(crypto_price)
        crypto_circulating_supply_list.append(crypto_circulating_supply)
        crypto_ticker_list.append(crypto_ticker)

#run the function with a nominated date
crypto_scrape()
```

```python
#store the data in a dataframe to make it structured
df['Name'] = crypto_name_list
df['Market cap'] = crypto_market_cap_list
df['Price'] = crypto_price_list
df['Circulating supply'] = crypto_circulating_supply_list
df['Ticker'] = crypto_ticker_list

#show the data
df
```

```python

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import time


# function to use Selenium to iteratively click next pages and store data daily and as a compiled list
def click_next_week():
    
    #code for clicking off notification alert (:2 is diable notifications)
    
    chrome_options = webdriver.ChromeOptions()
    prefs = {"profile.default_content_setting_values.notifications" : 2}
    chrome_options.add_experimental_option("prefs",prefs)
    driver = webdriver.Chrome(chrome_options=chrome_options)
    

   

    driver.get('https://coinmarketcap.com/historical/20130519/')

    time.sleep(2) # Delay to avoid looking like a bot to server
 

    #parsing html to Beautiful Soup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

   #creating empty lists with which to store one day's worth of data later
    crypto_name_list_one_day = []
    crypto_market_cap_list_one_day = []
    crypto_price_list_one_day = []
    crypto_circulating_supply_list_one_day = []
    crypto_ticker_list_one_day = []

#creating empty lists with which to store all of the daily data
    crypto_name_list = []
    crypto_market_cap_list = []
    crypto_price_list = []
    crypto_circulating_supply_list = []
    crypto_ticker_list = []
    
    



    num_pages = 5

  
    
    
    #page click counter
    page_click_counter = 0
    #creating while loop to click a certain number of pages
    while (page_click_counter < num_pages):
        count = 0
        next_week_button = driver.find_element(by=By.XPATH, value= '//*[@id="__next"]/div[1]/div[2]/div/div[1]/div[2]/div[2]/a[2]')
        
        next_week_button.click()
        
        print('button_clicked')

        

        #create a variable for the table rows (tr)
        tr = soup.find_all('tr', attrs={'class':'cmc-table-row'})
    #iterate through the first 10 rows
        
        while count < 10:
            
               
            
            # increase the counter for appending rows within 1 day
            
            for row in tr:
    #store the name of the coin as a variable by finding td element (column) 
                name_column = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__name'})
                #find the column within the row and strip the text
                crypto_name = name_column.find('a', attrs={'class': 'cmc-table__column-name--name cmc-link'}).text.strip()
                print(crypto_name)

                #find and createvvariable for market cap
                crypto_market_cap = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap'}).text.strip()
                print(crypto_market_cap)

                #Find and create variable for the price
                crypto_price = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price'}).text.strip()
                print(crypto_price)
                #Find and create variable for the circulating supply
                crypto_circulating_supply_and_ticker = row.find('td', attrs={'class': 'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply'}).text.strip()
                print(crypto_circulating_supply_and_ticker)
                #Find and create variable for the circulating supply by splitting the supply+ticker variable
                crypto_ticker = crypto_circulating_supply_and_ticker.split(' ')[1]
                print(crypto_ticker)
                #Find and create variable for the coin ticker
                crypto_circulating_supply = crypto_circulating_supply_and_ticker.split(' ')[0]
                print(crypto_circulating_supply)
                

                #append the data to the associated lists
                crypto_name_list_one_day.append(crypto_name)
                crypto_market_cap_list_one_day.append(crypto_market_cap)
                crypto_price_list_one_day.append(crypto_price)
                crypto_circulating_supply_list_one_day.append(crypto_circulating_supply)
                crypto_ticker_list_one_day.append(crypto_ticker)
                
                count = count +1
            

        crypto_name_list.append(crypto_name_list_one_day)
        
        crypto_market_cap_list.append(crypto_market_cap_list_one_day)
        
        crypto_price_list.append(crypto_price_list_one_day)
     
        crypto_circulating_supply_list.append(crypto_circulating_supply_list_one_day)
      
        crypto_ticker_list.append(crypto_ticker_list_one_day)
        

        next_week_button = driver.find_element(by=By.XPATH, value= '//*[@id="__next"]/div[1]/div[2]/div/div[1]/div[2]/div[2]/a[2]')
        next_week_button.click()
        
    
        page_click_counter += 1 
  
    print(crypto_name_list)
   
    print(crypto_market_cap_list)

    print(crypto_price_list)

    print(crypto_circulating_supply_list)
 
    print(crypto_ticker_list)
click_next_week()
```
